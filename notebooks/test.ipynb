```python
import requests
response = requests.get("https://openai.com/policies/terms-of-use")
response.encoding = 'utf-8'
```

```python
import re
from markdownify import MarkdownConverter

class IgnoreConverter(MarkdownConverter):
    """
    Create a custom MarkdownConverter that ignores headers and navs
    """

    TABOO_REGEX = [".*hidden.*", ".*header.*", ".*footer.*", ".*nav .*", ".*nav-.*"]
 
    def process_tag(self, node, convert_as_inline, children_only=False):
        for attr, value in node.attrs.items():
            if any(bool(re.match(regex, str(value).lower())) for regex in self.TABOO_REGEX):
                return ""
        return super().process_tag(node, convert_as_inline, children_only)
    def convert_header(self, el, text, convert_as_inline):
        return ''
    def convert_footer(self, el, text, convert_as_inline):
        return ''
    def convert_nav(self, el, text, convert_as_inline):
        return ''
    def convert_title(self, el, text, convert_as_inline):
        return ''
    def convert_button(self, el, text, convert_as_inline):
        return ''
    def convert_input(self, el, text, convert_as_inline):
        return ''
    def convert_form(self, el, text, convert_as_inline):
        return ''
    def convert_style(self, el, text, convert_as_inline):
        return ''
    def convert_iframe(self, el, text, convert_as_inline):
        return ''

def md(html, **options):
    return IgnoreConverter(**options).convert(html)

with open("out.md", "+w") as f:
    f.write(md(response.text))
    # f.write(md("\n".join([str(e) for e in res])))
```
